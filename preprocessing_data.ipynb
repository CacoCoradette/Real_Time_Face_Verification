```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import cv2
import os
```

```python
datagen = ImageDataGenerator(#width_shift_range=[-.5,+.5],
                            horizontal_flip=True,
                            vertical_flip=False,
                            # rotation_range=20, 
                            fill_mode='nearest',
#                             zoom_range=0.3,
                            brightness_range=[0.5,2.0]
                            )
```

```python
dir_names = os.listdir('friends')
names =  [dir_name.replace('.JPG','') for dir_name in dir_names]
names
```

```python
for name in names:
    
    img = cv2.imread('friends/{}.JPG'.format(name))
    img = cv2.resize(img,(96,96))
    data = img_to_array(img)
    samples = np.expand_dims(data, 0)
    
    it = datagen.flow(samples, batch_size=1)

    for i in range(1,4):
        image = next(it)[0].astype('uint8')
        cv2.imwrite('images_new/{}-{}.jpg'.format(name,i),image)
```

```python
dir_names = os.listdir('images_new')
names =  [dir_name.replace('.jpg','') for dir_name in dir_names]
for i,name in enumerate(names) :
    print(dir_names[i],name)
```
